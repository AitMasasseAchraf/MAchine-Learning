```python
fig=plt.figure(figsize=(25,25))
index=1 #The subplot will take the index position on a grid with nrows rows and ncols columns.
#index starts at 1 in the upper left corner and increases to the right
for i in range(1,5):
    for j in range(1,5):
        if i!=j:
            plt.subplot(4,3,index)
            plt.scatter(df.iloc[:,i],df.iloc[:,j],c=df["Species"])
            plt.xlabel(df.columns[i])
            plt.ylabel(df.columns[j])
            index=index+1
```

```python
from sklearn.neighbors import KNeighborsClassifier
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split


X=df.iloc[:,1:5].values

y=df["Species"].values
x_train,x_test,y_train,y_test=train_test_split(X,y)

knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(x_train[:, 2:], y_train)
plot_decision_regions(x_train[:, 2:], y_train[:], knn_model)
plt.xlabel('petal length[cm]')
plt.ylabel('petal width[cm]')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler() #z=x-u/l'ecart type in english is standard deviation 
scaler.fit(x_train)
x_train_std = scaler.transform(x_train)
x_test_std = scaler.transform(x_test)
# x_train_st
```

```python
df = pd.read_csv('C:\\Users\\aitma\\Downloads\\scikit-learn\\categorical.csv')
df
```

```python
mapping_dict = {'M': 2,
                'L': 3,
                'XXL': 5}

df['size'] = df['size'].map(mapping_dict)
df
```

```python
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder() # to encode text columns like labelEncoder
df['classlabel'] = le.fit_transform(df['classlabel'])
# np.unique(df["color"])
df
```

```python
from sklearn.compose import make_column_selector as selector

numerical_columns_selector = selector(dtype_exclude=object) # to select just all the columns except object columns
categorical_columns_selector = selector(dtype_include=object) # to select the object columns

numerical_columns = numerical_columns_selector(df)
categorical_columns = categorical_columns_selector(df)
# numerical_columns=["price"] because int64 has a specific number of bytes but string no because each string has its bytes numbers 
#so color column or is an array object which its items are string type.
# categorical_columns=["color"]
numerical_columns
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
#one hot encode your categorical data
#'ignore' : When an unknown category is encountered during transform,
#the resulting one-hot encoded columns for this feature will be all zeros. In the inverse 
categorical_preprocessor = OneHotEncoder(handle_unknown="ignore")
numerical_preprocessor = StandardScaler()
```

```python
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer([
    ('one-hot-encoder', categorical_preprocessor, categorical_columns),
    ('standard_scaler', numerical_preprocessor, numerical_columns)])
```

```python
preprocessor.fit(df)
```

```python
df2=preprocessor.transform(df)
```
