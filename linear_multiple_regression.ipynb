```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x1=np.array([1,1.5,2,2.5,3,4,4,5,5,5.5,6,6.5,7])
y=np.array([1,2.5,1,2.5,3.5,2.5,3,3,3.5,4.5,4.5,4,4.8])
x0=np.ones(13)
# x1,y,x0
x1=x1.reshape(13,1)
x0=x0.reshape(13,1)
y=y.reshape(13,1)
# x1,y,x0
```

```python
X=np.hstack((x1,x0))
print(X)
```

```python
plt.scatter(x1,y,c=y)
```

```python
 def h(X,theta):
    return np.dot(X,theta)
    
    
    
    
        
        



```

```python
theta=np.array([0,0]).reshape(2,1)
y_pred=h(X,theta)
print(y_pred)
```

```python
plt.scatter(x1,y)
plt.plot(x1,y_pred, c='r')
```

```python
def cout(X,theta,y):
    return np.sum((h(X,theta)-y)**2)/(2*len(X))

cout(X,theta,y)
```

```python
def  desc_gard(X,theta,y,pas):
    return theta-pas*np.dot(X.T,h(X,theta)-y)/len(X)
```

```python
def fit(X,theta,y,pas=0.01,n=100):
    cs=[]
    for i in range(n):
        cs.append(cout(X,theta,y))
        theta=desc_gard(X,theta,y,pas)
    return theta,cs    
```

```python
t,c=fit(X,theta,y,pas=0.01,n=100)
```

```python
t,c

```

```python
y_pred=h(X,t)
print(y_pred)

plt.scatter(x1,y)
plt.plot(x1,y_pred,c="r")
```
